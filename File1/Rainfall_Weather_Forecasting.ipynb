```python
def null_percentage(data):
    total_rows = len(data)
    null_percentage = (data.isnull().sum()/total_rows)*100
    print('Percentage of Null Values in each column:')
    for column, percentage in null_percentage.items():
        print(f'{column}:{percentage:.1f}%')
        
null_percentage(data)
```

```python
from sklearn.impute import SimpleImputer
def impute_null_val(data):
    num_imputer = SimpleImputer(strategy='mean')
    categ_imputer = SimpleImputer(strategy='most_frequent')
    num_cols = data.select_dtypes(include='number').columns
    categ_cols = data.select_dtypes(exclude='number').columns
    data[num_cols] = num_imputer.fit_transform(data[num_cols])
    data[categ_cols] = categ_imputer.fit_transform(data[categ_cols])
    return data

data = impute_null_val(data)
```

```python
data.drop(['Date','Location'],axis=1,inplace=True)
```

```python
#Creating a model for predicting whether or not it will rain tomorrow.
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
encoder = LabelEncoder()
cat_cols = ['WindGustDir','WindDir9am','WindDir3pm','RainToday']
data[cat_cols] = data[cat_cols].apply(encoder.fit_transform)
```

```python
x = data.drop('RainTomorrow',axis=1)
y = data['RainTomorrow']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
model = RandomForestClassifier(n_estimators=100,random_state=100)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Creating a model to predict how much rainfall could be there.
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
label_encoder = LabelEncoder()
data['RainTomorrow'] = label_encoder.fit_transform(data['RainTomorrow'])
```

```python
x = data.drop('Rainfall',axis=1)
y = data['Rainfall']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
model = RandomForestRegressor(n_estimators=100,random_state=100)
model.fit(x_train,y_train)
```

```python
# Making prediction and evaluating the model
y_pred = model.predict(x_test)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print(f"Mean Absolute Error: {mae:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-Squared: {r2:.2f}")
```

Precision: Precision measures the proportion of true positive predictions among all positive predictions made by the model.

Precision for "No" class: 0.91
Precision for "Yes" class: 0.85
These values indicate that when the model predicts "No rain" for a given day, it is correct approximately 91% of the time, and when it predicts "Rain" for a given day, it is correct approximately 85% of the time.

Recall: Recall measures the proportion of true positive predictions among all actual positive instances in the dataset.

Recall for "No" class: 0.97
Recall for "Yes" class: 0.65
These values indicate that the model correctly identifies approximately 97% of the actual instances where it did not rain ("No"), but only approximately 65% of the instances where it did rain ("Yes").

F1-score: The F1-score is the harmonic mean of precision and recall, providing a balance between the two metrics. It ranges from 0 to 1, where higher values indicate better model performance.

F1-score for "No" class: 0.94
F1-score for "Yes" class: 0.74
These values indicate that the model achieves a good balance between precision and recall for predicting "No rain" (F1-score of 0.94), but a lower balance for predicting "Rain" (F1-score of 0.74).

Mean Squared Error (MSE): This metric represents the average squared difference between the predicted values and the actual values. In this case, the MSE of 130.68 is relatively high, indicating that there are larger errors between the predicted and actual values, particularly when squared.

R-squared values range from 0 to 1, with higher values indicating better model fit. In this case, the R-squared value of 0.30 indicates that approximately 30% of the variance in the rainfall data is explained by the model.
