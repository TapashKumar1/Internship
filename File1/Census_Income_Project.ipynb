Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
categorical_cols = ['Workclass', 'Education', 'Marital_status', 'Occupation', 
                    'Relationship', 'Race', 'Sex', 'Native_country']
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])
    
print(data.head())
```

```python
# Checking for Skewness in the dataset 
features = data.drop('Income',axis=1)
skewness = features.apply(lambda x: x.skew())
print('Skewness for each features')
print(skewness)
```

```python
# Treating Skewness using log transformation
threshold = 1 
skewed_cols = {'Age': 0.558738,
    'Workclass': -0.752280,
    'Fnlwgt': 1.446972,
    'Education': -0.934063,
    'Education_num': -0.311630,
    'Marital_status': -0.013448,
    'Occupation': 0.114540,
    'Relationship': 0.786784,
    'Race': -2.435332,
    'Sex': -0.719244,
    'Capital_gain': 11.953690,
    'Capital_loss': 4.594549,
    'Hours_per_week': 0.227636,
    'Native_country': -3.658235
}

for col, skewness in skewed_cols.items():
    if abs(skewness) > threshold:
        data[col] = np.log1p(data[col])
```

Creating a model 

```python
# Importing the libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
```

```python
# Splitting the data into 'x' and 'y'
x = data.drop('Income',axis=1)
y = data['Income']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
# Model Evalusation
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:',accuracy)
print('\nClassification Report:')
print(classification_report(y_test,y_pred))
```

Creating a different model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 100, random_state = 100)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
# Model Evaluation 
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:',accuracy)
print('\nClassification Report:')
print(classification_report(y_test,y_pred))
```
