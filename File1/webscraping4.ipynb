1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url
= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 

A)Rank
 B) Name
 C) Artist
 D) Upload date
 E) Views 

```python
# Importing the necessary libraries for selenium, exception handling and pandas
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from time import sleep
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
# Navigating the website link using get method
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver = webdriver.Chrome()
driver.get(url)
sleep(2)
```

```python
# Finding the table element and scraping the data from the table.
table = driver.find_element(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]')
table
```

```python
# Finding the row element and scraping the data from all the rows.
rows = table.find_elements(By.TAG_NAME,'tr')
rows
```

```python
# Creating an empty list named data, looping through the rows using for loop and finding the columns, 
# while scraping the column data using index numbers.
data = []
for row in rows[1:]:
    column = row.find_elements(By.TAG_NAME,"td")
    if len(column)>=5:
        rank = column[0].text
        name = column[1].text
        uploader = column[2].text
        views = column[3].text
        date = column[4].text
    
        data.append({
            'Rank':rank,
            'Name':name,
            'Uploader':uploader,
            'Views':views,
            'Date':date,
        })
```

```python
# Creating a pandas dataframe
df = pd.DataFrame(data)
```

```python
df
```

```python
# Saving the dataframe to a csv
df.to_csv("most_viewed_yt_videos.csv",index=False)
```

```python
driver.quit()
```
