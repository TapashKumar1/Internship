Summary of the correlation matrix:-

1. Happiness Rank vs Other Features:- Hapinness rank column has a strong negative relation with happiness score and economy. It also has negative correlations with other features such as family, Health, Freedom.

2. Happiness Score vs other features:- Happiness score is highly positevely correlated with Economy and positively correlated with family, heath and freedom

3. Standard Score vs other features:- Standard score has a very weak negative correlation with most of the features except for itself which is expected since its the same variable.

4. Economy vs other features:- Economy has a strong negative correlation with happiness rank and strong positive correlation with happiness score and health life

5. Family vs other variables:- Family has a negative correltaion with happiness rank and postive correlation with happiness score

6. Health vs other variables:- Health has a negative correltaion with happiness rank and postive correlation with happiness score

7. Freedom vs other variables:- Freedom has a negative correltaion with happiness rank and postive correlation with happiness score and weak relation with other features

8. Trust vs other features:- Trust has weak correlations with other variables

9. Generosity vs other features:- Generosity has weak correlations with other variables

10. Dystopia vs other features:- Dystopia has postive correlation with happiness score and negative correlation with happiness rank with fairly weak relations with other variables.
