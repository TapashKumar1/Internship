```python
# Treating class imbalance with RandomOverSampler and RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter

x = data.drop(columns=['Attrition'])
y = data['Attrition']

print("Class Distribution before balancing", Counter(y))

over_sampler = RandomOverSampler(sampling_strategy='minority')
under_sampler = RandomUnderSampler(sampling_strategy='majority')

x_resampled_over, y_resampled_over = over_sampler.fit_resample(x,y)
x_resampled,y_resampled = under_sampler.fit_resample(x_resampled_over,y_resampled_over)

print("Class distribution after balancing:", Counter(y_resampled))
```

```python
data = x_resampled.copy()
data['Attrition']=y_resampled
```

```python
# Plotting target variable after treating class imbalance
attrition_percentage = data['Attrition'].value_counts(normalize=True)*100
plt.figure(figsize=(6,4))
sns.barplot(x=attrition_percentage.index,y=attrition_percentage.values,palette='viridis')
plt.title('Percentage of Values in Attrition Caolumn')
plt.xlabel('Attrition')
plt.ylabel('Percentage')
plt.show()
```

```python
# Changing the Categorical columns into Numerical using One Hot encoding 
categorical_cols = data.select_dtypes(include=['object']).columns
data_encoded = pd.get_dummies(data,columns=categorical_cols,drop_first=True)
print(data_encoded.head())
```

```python
# Checking for skewness in the data
data = data_encoded
data.skew()
```

```python
# Applying log transformation and square root transformation to treat skewness in positively skewed features and negatively skewed features
positive_skew_vars = ['JobLevel', 'MonthlyIncome', 'YearsAtCompany', 'YearsSinceLastPromotion',
                      'NumCompaniesWorked', 'PercentSalaryHike', 'PerformanceRating',
                      'StockOptionLevel', 'TotalWorkingYears', 'YearsInCurrentRole',
                      'YearsWithCurrManager', 'BusinessTravel_Travel_Frequently',
                      'Department_Sales', 'EducationField_Marketing', 'EducationField_Medical',
                      'EducationField_Other', 'EducationField_Technical Degree',
                      'JobRole_Human Resources', 'JobRole_Laboratory Technician',
                      'JobRole_Manager', 'JobRole_Manufacturing Director',
                      'JobRole_Research Director', 'JobRole_Research Scientist',
                      'JobRole_Sales Executive', 'JobRole_Sales Representative',
                      'MaritalStatus_Single', 'OverTime_Yes']

negative_skew_vars = ['DistanceFromHome', 'JobInvolvement', 'WorkLifeBalance',
                      'BusinessTravel_Travel_Rarely', 'Department_Research & Development',
                      'EducationField_Life Sciences', 'Gender_Male', 'MaritalStatus_Married']

data[positive_skew_vars] = np.log1p(data[positive_skew_vars])
data[negative_skew_vars] = np.sqrt(data[negative_skew_vars])
print(data.head())
```

Creating a Classification Model

Random Forest Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#Splitting the data into features (X) and target variable (y) and Splitting the data into training and testing sets
x = data.drop('Attrition_Yes',axis=1)
y = data['Attrition_Yes']
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=100)
```

```python
# Initialize and train the Random Forest classifier
model = RandomForestClassifier(random_state=100)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
# Evaluating Moels perfromance
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy:',accuracy)
print("\nClassification Report:")
print(classification_report(y_test,y_pred))
```

Conclusion:-

The model achieved an accuracy of approximately 98.58%, indicating that it correctly classified the instances into their respective classes with high precision.

Precision, Recall, and F1-score: 

The classification report shows high precision, recall, and F1-score values for both classes (False and True). Precision measures the proportion of true positives among all instances predicted as positive, while recall measures the proportion of true positives that were correctly predicted. The F1-score is the harmonic mean of precision and recall, providing a balance between the two metrics. The high values of these metrics indicate that the model performs well in classifying both classes.

Support: 

The support values represent the number of occurrences of each class in the testing dataset. The support values for both classes are balanced, with 249 instances of the False class and 245 instances of the True class.
